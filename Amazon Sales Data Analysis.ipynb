Which region has the highest total sales revenue?

```python
df['Highest Sales Region'] = df.groupby('Region')['Total Revenue'].sum().idxmax()
print("Region with highest total sales revenue:", df['Highest Sales Region'].iloc[0])
```

```python
x=np.array(df['Region'].unique())
y=np.array(df.groupby('Region')['Total Revenue'].sum())
plt.figure(figsize=(10,6))
plt.plot(x, y, color='skyblue')
plt.xticks(rotation=45)
sns.set_style('darkgrid')
plt.title('Total Sales Revenue by Region')
plt.xlabel('Region',fontsize=14)    
plt.ylabel('Total Sales Revenue')
plt.show()
```

```python
my_labels = df['Region'].unique()
my_values = df.groupby('Region')['Total Revenue'].sum() 

plt.pie(my_values, labels=my_labels,autopct='%1.1f%%', startangle=140,textprops={'fontsize': 9, 'color': 'black', 'fontfamily': 'Arial'})
plt.title('Sales Revenue Distribution by Region')
plt.show()
```

Which item type have the highest & lowest total sales?


```python
revenue_sales=df.groupby('Item Type')['Total Revenue'].sum()


Highest_revenue=revenue_sales.idxmax()
Lowest_revenue=revenue_sales.idxmin()    
Highest_sales_value=revenue_sales.max()
Lowest_sales_value=revenue_sales.min()  
print("Item Type with Highest Sales Revenue:\n", Highest_revenue, "=", Highest_sales_value)
print("Item Type with Lowest Sales Revenue:\n", Lowest_revenue, "=", Lowest_sales_value)

count_sales=df.groupby('Item Type')['Order ID'].count()
Highest_sales=count_sales.idxmax()
Lowest_sales=count_sales.idxmin()   
Highest_sales_count=count_sales.max()
Lowest_sales_count=count_sales.min()    
print("Item Type with Highest Number of Sales:\n", Highest_sales, "=", Highest_sales_count)
print("Item Type with Lowest Number of Sales:\n", Lowest_sales, "=", Lowest_sales_count)
```

Are there any seasonal trends & patterns in sales data?

```python
df['Month'] = df['Order Date'].dt.month
import calendar as cal
df['month_names'] = df['Month'].apply(lambda x: cal.month_name[x])
Sales = df.groupby(['Month','month_names'])['Total Revenue'].sum().reset_index()
Sales=Sales.sort_values('Month')
Sales

```

```python
import calendar as cal


Month = df['Order Date'].dt.month
Sales = df.groupby(Month)['Total Revenue'].sum()


Month = np.array(Sales.index)
Sales = np.array(Sales.values)

month_names = [cal.month_name[i] for i in Month]    

plt.figure(figsize=(10,6))
plt.bar(month_names, Sales, color='green')
plt.xticks(month_names, rotation=45)

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()

#1e7 is a scientific form which means approx 10 million

```

Are there any outliers in the total cost distribution?

```python
Q1= df['Total Cost'].quantile(0.25)
Q3= df['Total Cost'].quantile(0.75)
IQR= Q3 - Q1
lower_bound= Q1 - 1.5 * IQR
upper_bound= Q3 + 1.5 * IQR
outliers = df[(df['Total Cost'] < lower_bound) | (df['Total Cost'] > upper_bound)]
outliers[['Order ID','Total Cost']]
```

How does the order priority vary based on the item type?

Item Type, H priority, M Priority, L Priority , C Priority, Total

```python
Item_priority=df.groupby('Item Type').agg(H_priority=('Order Priority', lambda x: (x=='H').sum()),
                                        M_priority=('Order Priority', lambda x: (x=='M').sum()),    
                                        L_priority=('Order Priority', lambda x: (x=='L').sum()),    
                                        C_priority=('Order Priority', lambda x: (x=='C').sum()))
Item_priority['Total']=Item_priority.sum(axis=1)
Item_priority.loc['Total']=Item_priority.sum(numeric_only=True, axis=0)
Item_priority
```
